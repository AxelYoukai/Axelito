```python
def Leer(Documento):
    import pandas as pd
    import numpy as np
    try:
        dfc = pd.read_csv(Documento,sep=',',encoding="utf-8")
    except:
        dfc = pd.read_excel(Documento,sep=',',encoding="utf-8")
    return dfc
DatasetS=Leer('comentarios-etiquetados_CC_UTF-8.csv');
DatasetS
```

Función para clasificación - Mappeo

```python
def clasificar(dfc):
    Comentario={}
    for indice in dfc['X']: #Por cada dato
        try:
            Comentario[dfc['Criterio'][int(indice)-1]].append(dfc['Comentario'][int(indice)-1]) #Se almacena en el criterio que refleja el dataset
        except: #Si hay error es porque no se ha inicializado la variable con su key
            Comentario[dfc['Criterio'][int(indice)-1]]=[dfc['Comentario'][int(indice)-1]] #Se inicializa mientras se almacena en vector
    return Comentario

Comentarios=clasificar(DatasetS)
del DatasetS
Comentarios
```

```python
def nlpito():
    import spacy    
    nlp = spacy.load("es_core_news_md")
    Datos=Leer('PSE.csv')
    for e in Datos:
        nlp.vocab[e].is_stop = False
        nlp.vocab[e.lower()].is_stop = False
    nlp.vocab["me"].is_stop = False
    return nlp
```

Función para stemmer

```python
def StemComentarios(Comentarios):
    import nltk
    from nltk.tokenize import word_tokenize
    from nltk.stem import SnowballStemmer
    Resultado={}
    StemS=SnowballStemmer('spanish')
    for k in Comentarios.keys(): #Por cada criterio en la lista de comentarios
        Tokenizado = [word_tokenize(comentario.lower(),"spanish") for comentario in Comentarios[k]] ##Por cada comentario dentro de criterio indice
        Resultado[k]=[]
        for ComentarioT in Tokenizado:
            ComentarioT= [StemS.stem(Token.lower()) for Token in ComentarioT]
            Sentencia=' '.join(ComentarioT)
            Resultado[k].append(Sentencia)
        del Tokenizado
    return Resultado

#Comentarios= StemComentarios(Comentarios)
#Comentarios
    
```
